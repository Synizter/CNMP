```python
import numpy as np
from scipy.signal import coherence
import matplotlib.pyplot as plt
fs = 500
f = 1
t = np.linspace(0,5, fs)
x = np.sin(2 * np.pi * f * t) + np.sin(2 * np.pi * (f + 12) * t) + np.sin(2 * np.pi * (f + 3) * t)
y = np.sin(2 * np.pi * f * t)

plt.plot(t,x, label = 'X')
plt.plot(t, y, label = 'Y')

plt.grid()
plt.legend()
plt.show()
```

```python
from pywt import cwt

scales = np.arange(1,41, 1)

coeff_x, freq_x = cwt(x, scales, 'mexh')
coeff_y, freq_y = cwt(y, scales, 'mexh')

f, axs = plt.subplots(2,2, figsize = (12,5))
axs[0][0].plot(t,x)
axs[1][0].imshow(np.abs(coeff_x), extent=[0, 1, 1, 41], aspect ='auto', cmap='jet', interpolation = 'bilinear')
axs[0][1].plot(t,y)
axs[1][1].imshow(np.abs(coeff_y), extent=[0, 1, 1, 41], aspect ='auto', cmap='jet', interpolation = 'bilinear')
plt.show()
```

```python
freq, coh = coherence(coeff_x, coeff_y, fs = fs)
print(coh.mean())
plt.semilogy(freq, coh.mean(0))
plt.grid()
plt.show()
```
